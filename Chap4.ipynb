```python
X = mnist.train.images
Y = mnist.train.labels
print(X.shape)
print(Y.shape)

n = X.shape[0] #画像の枚数
N = 100 #モデルの学習及び評価で使う枚数
Class = Y.shape[1] #クラス数（分類）
indices = np.random.permutation(range(n))[:N] #ランダムにN枚選択。

X_N = X[indices]
Y_N = Y[indices]

#train:test = 8:2でランダムに分割。
X_train, X_test, Y_train, Y_test = train_test_split(X_N, Y_N, train_size = 0.8) 

feature_size = X_N.shape[1] #特徴量の大きさ
neuron_size = 200 #隠れ層のニューロンの数
```

```python
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.optimizers import SGD
```

```python
# model(Keras)
model = Sequential() #モデルの受け皿

model.add(Dense(neuron_size, input_dim = feature_size))#層の追加(ニューロンの数、インプットの数を入力)
model.add(Activation('tanh'))                          #活性化関数

model.add(Dense(neuron_size))#層の追加(ニューロンの数入力)
model.add(Activation('tanh'))#活性化関数

model.add(Dense(neuron_size))#層の追加(ニューロンの数入力)
model.add(Activation('tanh'))#活性化関数

model.add(Dense(Class))         #層の追加(出力：0~9の分類)
model.add(Activation('softmax'))#活性化関数

model.compile(loss = "categorical_crossentropy", #損失関数の決定
              optimizer = SGD(lr = 0.01),         #最適化の方法(確率的勾配降下法に決めている)
              metrics = ['accuracy']             #評価方法の決定
             )
```

```python
%%time
#modelの学習
epochs = 1000
batch_size = 100

model.fit(X_train, Y_train, epochs = epochs, batch_size = batch_size)
```

```python
#評価
loss_and_metrics = model.evaluate(X_test, Y_test)
print(loss_and_metrics) #損失関数(目的関数)の値、metric(compileで決定)
```
